## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
crimes_recorded = pd.read_sql("SELECT DISTINCT COUNT(*) AS total_num_of_crime FROM CHICAGO_CRIME_DATA", conn)
crimes_recorded
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
community_areas_income_less_than_11000 = pd.read_sql("SELECT COMMUNITYAREANAME AS areas_with_income_less_than_11000 FROM CENSUS_DATA WHERE PERCAPITAINCOME < 11000 ", conn)
community_areas_income_less_than_11000
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
crimes_involving_minors = pd.read_sql("SELECT CaseNumber AS case_nums_involving_minors FROM CHICAGO_CRIME_DATA WHERE Description LIKE '%MINOR%'", conn)
crimes_involving_minors
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
child_kidnappings = pd.read_sql("SELECT * FROM CHICAGO_CRIME_DATA WHERE primarytype = 'KIDNAPPING' AND description LIKE 'CHILD%'", conn)
child_kidnappings
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
crimes_at_shcools = pd.read_sql("SELECT DISTINCT primarytype AS desc_of_school_crimes FROM CHICAGO_CRIME_DATA WHERE LocationDescription LIKE '%school%'", conn)
crimes_at_shcools
```

### Problem 6

##### List the average safety score for each type of school.


```python
avg_safety_score = pd.read_sql("SELECT ( CASE WHEN ElementaryMiddleorHighSchool = 'ES' THEN 'Elementary School' WHEN ElementaryMiddleorHighSchool = 'MS' THEN 'Middle School' WHEN ElementaryMiddleorHighSchool = 'HS' THEN 'High School' END) AS School_Type, AVG(safetyscore) AS avg_safety_score FROM CHICAGO_PUBLIC_SCHOOLS GROUP BY 1", conn)
avg_safety_score
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
top_5_communities_below_poverty_line = pd.read_sql("SELECT COMMUNITYAREANAME AS community_areas, PERCENTHOUSEHOLDSBELOWPOVERTY AS percent_below_poverty_line FROM CENSUS_DATA GROUP BY 1 ORDER BY 2 DESC LIMIT 5", conn)
top_5_communities_below_poverty_line
```

### Problem 8

##### Which community area is most crime prone?


```python
most_crime_area = pd.read_sql("SELECT CommunityAreaName AS area_name, CommunityArea AS community_area_num, COUNT(CommunityArea) AS total_crime_count FROM CENSUS_DATA cd JOIN CHICAGO_CRIME_DATA ccd ON cd.CommunityAreaNumber = ccd.CommunityArea GROUP BY 2 ORDER BY 3 DESC LIMIT 1", conn)
most_crime_area 
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
highest_hardship = pd.read_sql("SELECT communityareaname FROM census_data where communityareanumber = (SELECT communityareanumber FROM CENSUS_DATA GROUP BY 1 ORDER BY hardshipindex DESC limit 1)", conn)
highest_hardship 

```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
most_crime_area_name = pd.read_sql("SELECT area_name AS community_with_most_crime, MAX(total_crime_count) AS crime_count FROM (SELECT CommunityAreaName AS area_name, CommunityArea AS community_area_num, COUNT(CommunityArea) AS total_crime_count FROM CENSUS_DATA cd JOIN CHICAGO_CRIME_DATA ccd ON cd.CommunityAreaNumber = ccd.CommunityArea GROUP BY 2 ORDER BY 3) AS t1 ", conn)
most_crime_area_name 


#SELECT communityareaname FROM CENSUS_DATA WHERE communityareanumber =
```
